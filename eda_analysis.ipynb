EDA(Exploratory Data Analysis)

```python
import pandas as pd             # Used to read and manipulate data.
import matplotlib.pyplot as plt # A library for creating static, animated, and interactive visualizations.
import seaborn as sns           # A higher-level interface for drawing attractive and informative statistical graphics.
```

```python
input_file = "D:\\Omnify\\cleaned_data.csv"
df = pd.read_csv(input_file)  
```

```python
# Set Seaborn plot style to "whitegrid" for better readability
sns.set_style("whitegrid")
```

Booking Type Distribution


```python
# Booking Type Distribution (Fixed)
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x="Booking Type", order=df["Booking Type"].value_counts().index, hue="Booking Type", palette="viridis", legend=False)
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=10, color='black')
plt.xticks(rotation=45)
plt.title("Booking Type Distribution")
plt.xlabel("Booking Type")
plt.ylabel("Count")
plt.show()
```

Revenue by Booking Type


```python
# Revenue by Booking Type (No Fix Needed)
plt.figure(figsize=(6, 4))
ax = df.groupby("Booking Type")["Price"].sum().sort_values().plot(kind="barh", color="steelblue") 

for i, v in enumerate(df.groupby("Booking Type")["Price"].sum().sort_values()):
    plt.text(v + 50, i, str(int(v)), color='black', va='center', fontsize=10)  # Adjust the position if needed

plt.title("Total Revenue by Booking Type")
plt.xlabel("Total Revenue ($)")
plt.ylabel("Booking Type")
plt.show()
```

Status Distribution

```python
# Status Distribution (Fixed)
plt.figure(figsize=(6, 4))
ax = sns.countplot(data=df, x="Status", order=df["Status"].value_counts().index, hue="Status", palette="pastel", legend=False)
# Add value labels to bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', 
                (p.get_x() + p.get_width() / 2, p.get_height()), 
                ha='center', va='bottom', fontsize=10, color='black')
plt.title("Booking Status Distribution")
plt.xlabel("Status")
plt.ylabel("Count")
plt.show()
```

Peak Booking Time Analysis

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x="Time Slot", order=df["Time Slot"].value_counts().index, hue="Time Slot", palette="viridis",legend=False)
plt.title("Peak Booking Times")
plt.xticks(rotation=45)
plt.show()
```
